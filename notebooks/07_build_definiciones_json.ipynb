```python
from pathlib import Path

# Ruta al archivo markdown de definiciones (desde notebooks/)
md_path = Path("../data/raw/web/02_definiciones.md")

# Leer contenido completo
text = md_path.read_text(encoding="utf-8")

# Mostrar las primeras 10 líneas para comprobar que se ha cargado bien
text.splitlines()[:10]

```

```python
from pathlib import Path
import re
import json

# Ya tenemos 'text' de la celda anterior
lines = text.splitlines()

# Patrón para headings tipo: "## 2.1 Adjudicatario"
pat_heading = re.compile(r"^##\s+(2\.\d+)\s+(.+)$")

definiciones = []   # aquí guardaremos el resultado final
current = None      # definición actual (id + término)
buffer = []         # líneas de texto de la definición actual

for line in lines:
    m = pat_heading.match(line)
    if m:
        # Si ya estábamos en una definición anterior, la cerramos
        if current is not None:
            definicion_texto = " ".join(buffer).strip()
            definiciones.append({
                "id": current["id"],
                "termino": current["termino"],
                "definicion": definicion_texto
            })
            buffer = []

        # Creamos una nueva definición
        numero = m.group(1)          # "2.1", "2.2", ...
        termino = m.group(2).strip() # "Adjudicatario", etc.
        current = {
            "id": numero,
            "termino": termino
        }
    else:
        # Si no es un heading y ya hay una definición en curso,
        # añadimos la línea al buffer de texto.
        if current is not None:
            if line.strip():          # ignorar líneas vacías
                buffer.append(line.strip())

# Al salir del bucle, falta guardar la última definición
if current is not None:
    definicion_texto = " ".join(buffer).strip()
    definiciones.append({
        "id": current["id"],
        "termino": current["termino"],
        "definicion": definicion_texto
    })

len(definiciones), definiciones[0]
```

```python
from pathlib import Path
import json

json_path = Path("../data/raw/json/definiciones.json")
json_path.parent.mkdir(parents=True, exist_ok=True)

with json_path.open("w", encoding="utf-8") as f:
    json.dump(definiciones, f, ensure_ascii=False, indent=2)

json_path, len(definiciones)
```
